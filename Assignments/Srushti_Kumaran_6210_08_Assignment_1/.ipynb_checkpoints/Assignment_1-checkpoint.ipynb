# ◊ Source 1 - Web API

## ◊ Importing libraries:

```python
import requests
import pandas as pd
import json
import os
```

## ◊ Importing Data using API:

```python
api_key = 'aea0b4c3a78233428f8123142dcd2c63'
```

## ◊ Extracting Popular TV shows through TMDB API:

```python
# First initiate two empty list for storing the raw data
rawPopularTvShowJsonData = []
popularTvShow = []

# Merged the link and the API key to visit the full link
getPopularTvShowRequest = 'https://api.themoviedb.org/3/tv/popular?api_key='+ api_key +'&language=en-US&page='
print(getPopularTvShowRequest)
```

```python
# Iterating over the 20 pages and fetching the results in json format for the Popular TV Shows
for page in range(1,21):
    tvShowData = requests.get(getPopularTvShowRequest + str(page))
    rawPopularTvShowJsonData.append(tvShowData.json())
    popularTvShow += tvShowData.json()['results']
```

### ◊ Printing the generated data "tvShowData" in json format:

### Information on below bar graph is:

# ◊ Source 3 - Web Scraping using Beautiful Soup:  

### Beautiful Soup : 

Beautiful Soup is a Python library for pulling data out of HTML and XML files. It works with  parser to provide idiomatic ways of navigating, searching, and modifying the parse tree. It commonly saves programmers hours or days of work.

```python
# importing liabraries for web scraping
import requests
from bs4 import BeautifulSoup
```

```python
# Fetching the tags from the website
url = requests.get("https://www.themoviedb.org/tv", headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
response = url.content
print(response)
```

#### ◊ Displaying the above format to more readable format using html parser:

```python
#formatting above data
soup=BeautifulSoup(response,"html.parser")
print(soup.prettify)
```

#### ◊ Retrieving multiple attribute values and Printing:

```python
all=soup.find_all("div",{"class":"image_content"})
all[5]
```

```python
# Iterated the values
for i in all:
    i = soup.find("div",{"class":"image_content"}).find("a",{"class":"result"})["title"]
```

```python
# Fetched the Retrieved value in items
items = soup.find_all("div",{"class":"image_content"})
```

```python
# Printing items
items
```

```python
# Iterated in items and fetched and displayed only the tiltles of the films
for item in items:
    print(item.find("a",{"class":"result"})["title"].capitalize())
```

## ◊ Creating Tables and Combining them to get New table:

```python
#Creating Tv_Show_By_Popularity
Tv_Show_By_Popularity = [ 'id','name','popularity', 'genre_ids']                                                           
Tv_Show_By_Popularity = pd.DataFrame(popularTvShow, columns = Tv_Show_By_Popularity)                                              
Tv_Show_By_Popularity
```

```python
#Creating Tv_Show_Details
Tv_Show_Details= ['id', 'name', 'original_name', 'overview', 'first_air_date', 'original_language','origin_country', 'vote_average', 'vote_count']                         
Tv_Show_Details = pd.DataFrame(popularTvShow, columns = Tv_Show_Details)                                                    
Tv_Show_Details
```

```python
# Merging Tv_Show_Details table and Tv_Show_By_Popularity table
Tv_Show_Data_Combined = pd.merge(Tv_Show_Details, Tv_Show_By_Popularity,left_on='id',right_on='id')
Tv_Show_Data_Combined = Tv_Show_Data_Combined.drop_duplicates(subset='id', keep="first")
Tv_Show_Data_Combined
```

```python
#Generating csv file for combined database
Tv_Show_Data_Combined.to_csv("Tv_Show_Data_Combined.csv",encoding="utf-8",index=False)
Tv_Show_Data_Combined = pd.read_csv("Tv_Show_Data_Combined.csv")
Tv_Show_Data_Combined.head()
```

### ◊ CONCLUSION:  
We understood the concept of Data munging that it is the general procedure for transforming data from erroneous or unusable forms, into useful and use-case-specific ones. Without some degree of munging, whether performed by automated systems or specialized users, data cannot be ready for any kind of downstream consumption. 

We also learnt that Data auditing is the process of conducting a data audit to assess how data is fit for given purpose. This involves profiling the data and assessing the impact of poor quality data on the organization's performance and profits. It can include the determination of the clarity of the data sources and can be applied in the way organization and rating agencies perform due diligence with regard to the treatment of raw data given by firms, particularly the identification of faulty data.

Below are few point which enabled us to understand why auditing is mandatory:  
-Ensures the end-to-end integrity of data activities by identifying when modifications are made.  
-Detects and analyzes intentional and accidental breaches in user behavior.  
-Monitors and analyze the database activities of any user.  
-Keeps track of changes and updates made to data.  

In the process, we got familiar to conceptual database model which is defined as a structured view of the data required to support a processes, record events, and track related performance measures. This model focuses on identifying the data used in the business but not its processing flow or physical characteristics.

Also we got acquianted with entity relationship diagram (ERD), also known as an entity relationship model, which is a graphical representation that depicts relationships among people, objects, places, concepts or events (data) within a system. Entity relationship diagrams provide a visual starting point for database design that can also be used to help determine information system requirements throughout an organization.
